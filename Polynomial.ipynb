```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('https://raw.githubusercontent.com/Samrudhig/datasets/main/Polynomial-Dataset%20(1).csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
# Importing the Linear Regression class
regressor = LinearRegression()
# Fitting the linear regression model
regressor.fit(X, y)
```

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
regressor_2 = LinearRegression()
regressor_2.fit(X_poly, y)
```

```python
# Visualising the Linear Regression results
plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X), color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Polynomial Regression results
plt.scatter(X, y, color='red')
plt.plot(X, regressor_2.predict(poly_reg.fit_transform(X)), color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Fitting Polynomial Regression to the dataset with degree=3
poly_reg = PolynomialFeatures(degree=3)
# Creating polynomial features
X_poly = poly_reg.fit_transform(X)
# Fitting polynomial Regression
regressor_3 = LinearRegression()
regressor_3.fit(X_poly, y)
```

```python
# Visualising the Polynomial Regression results with degree=3
plt.scatter(X, y, color='red')
plt.plot(X, regressor_3.predict(poly_reg.fit_transform(X)), color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Fitting Polynomial Regression to the dataset with degree=4
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
regressor_4 = LinearRegression()
regressor_4.fit(X_poly, y)
```

```python
# Visualising the Polynomial Regression results with degree=4
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor_4.predict(poly_reg.fit_transform(X_grid)), color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting a new result with Linear Regression
level_of_employee = 6.5
salary_of_employee_linear = regressor.predict([[level_of_employee]])
print('new result with linear regression :', salary_of_employee_linear)
```

```python
# Predicting a new result with Polynomial Regression
salary_of_employee_polynomial = regressor_4.predict(poly_reg.fit_transform([[level_of_employee]]))
print('new result with polynomial regression :', salary_of_employee_polynomial)
```
